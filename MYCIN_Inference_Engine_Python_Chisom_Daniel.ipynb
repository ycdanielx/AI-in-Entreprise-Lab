```python
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(-1,1,0.25)
y=np.arange(-1,1,0.25)

def Mycin(x, y):
    if x > 0 and y > 0:
        return x + y - x * y
    elif x < 0 and y < 0:
        return x + y + x * y
    elif abs(x)==1 and abs(y)==1:
        return 1
    else:
        return (x + y) / (1 - min(abs(x), abs(y)))
```

```python
fig = plt.figure()
ax= plt.axes(projection='3d')
for i in x:
    for j in y:
        k = Mycin(i,j)
        ax.scatter(i,j,k, c='green')
plt.show()
```
